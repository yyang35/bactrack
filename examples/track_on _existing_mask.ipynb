## Run tracking on existing mask
- Read in the masks folder (in this case you don't need raw data folder)
- Do Tracking

Reading mask in below

```python
from bactrack.widget import get_hierarchies_from_masks_folder
mask_dir = "/Users/sherryyang/Documents/wiggins-lab/debug_phase"
hier_arr = get_hierarchies_from_masks_folder(mask_dir)
```

```python
from bactrack.tracking import OverlapWeight, IOUWeight, DistanceWeight
w =  OverlapWeight(hier_arr)
```

```python
from bactrack.tracking import MIPSolver, ScipySolver
solver = ScipySolver(w.weight_matrix, hier_arr)
n, e = solver.solve()
```

```python
from bactrack import io
masks, edge_df = io.format_output(hier_arr, n, e, overwrite_mask=False)
node_df = io.hiers_to_df(hier_arr)
```

### Make a master table for all information contains details links information

```python
import pandas as pd
hier_df = io.hiers_to_df(hier_arr)
merged_df = pd.merge(edge_df, hier_df.add_suffix('_source'), left_on='Source Index', right_on='index_source', how='left')
merged_df = pd.merge(merged_df, hier_df.add_suffix('_target'), left_on='Target Index', right_on='index_target', how='left')
```
