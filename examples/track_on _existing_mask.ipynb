## Run tracking on existing mask
- Read in the masks folder (in this case you don't need raw data folder)
- Do Tracking

Reading mask in below

```python
from bactrack.widget import get_hierarchies_from_masks_folder
mask_dir = "/Users/sherryyang/Documents/wiggins-lab/data/1106/xy5/masks"
hier_arr = get_hierarchies_from_masks_folder(mask_dir)
```

```python
from bactrack.tracking import OverlapWeight, IOUWeight, DistanceWeight
w =  DistanceWeight(hier_arr)
```

```python
from bactrack.tracking import MIPSolver, ScipySolver
```

```python
solver = ScipySolver(w.weight_matrix, hier_arr)
n, e = solver.solve()
```

```python
from bactrack import io
masks, edge_df = io.format_output(hier_arr, n, e)
node_df = io.hiers_to_df(hier_arr)
```
